```python
import pysentiment2 as ps
import csv

with open('dataset_2021.csv') as csv_file:
  csv_reader = csv.reader(csv_file, delimiter=',')
  line_count = 0
  pol = 0
  for row in csv_reader:
    if line_count == 0:
      print("column names:", row)
      line_count +=1
    if row[3] == "W":      
      text = row[2]
      hiv4 = ps.HIV4()
      tokens = hiv4.tokenize(text)
      print(tokens)
      score = hiv4.get_score(tokens)
      #print(f"{row[0]} sentiment: {score['Polarity']}.")
      line_count +=1
      pol += score['Polarity']


print(f"Average Polarity White Athletes: {pol/line_count}.")
```

```python
import numpy as np
import pandas as pd
import re
import requests
import time
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from requests import get
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
import csv
from urlextract import URLExtract

extractor = URLExtract()

def get_links():

  years = [2021]
  name_list = []
  position_list = []
  report_links = []
  count = 0
  for i in range(1):
    for x in range(1,8):
      y = years[i]
      url = "https://www.nfl.com/draft/tracker/picks/?round={0}&year={1}".format(x, y)
      driver.get(url)
      time.sleep(5)
      html = driver.page_source
      soup = BeautifulSoup(html, "html.parser")
      name = soup.findAll("a", attrs={"class": "css-ytumd6"})
      print(name)
      position = soup.findAll("div", attrs={"class": "css-1g96t07"})
      #print(position)
      round_size = len(position)+1
      print(round_size)
      for j in range(1,round_size):
        pick = j + count
        print(pick)
        url = "https://www.nfl.com/draft/tracker/picks/?pick={0}&round={1}&year={2}".format(pick, x, y)
        driver.get(url)
        time.sleep(5)
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        print(name[j-1])
        player_link = extractor.find_urls(str(name[j-1]))
        report_links.append(player_link)
        #print(report_links)
        name_stripped = name[j-1].get_text()
        name_list.append(name_stripped)
        pos_stripped = position[j-1].get_text()
        position_list.append(pos_stripped)
      count = count + round_size  

  return name_list, position_list, report_links

names,pos,report_links = get_links()
```
