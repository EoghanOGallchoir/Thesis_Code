```python
# read in GI as key-value pairs with keys being preprocessed (famili, confid, unnecessari, etc)
import csv
import pysentiment2 as ps
import re

def read_GI(filename):
  kvp = {}
  with open(filename ,'r', ) as file:
    csvreader = csv.reader(file)
    next(csvreader)
    next(csvreader)
    for col in csvreader:
      if col[1] != "Lvd": 
        kvp[col[0]] = [str(word) for word in col[2:-2] if word]
  
  kvp2 = {}
  hiv4 = ps.HIV4()   
  for key in kvp:
    new_key = str(hiv4.tokenize(key))
    # print(new_key)
    #print(bool(new_key))
    res = re.findall(r'\w+', new_key)
    # print(res)
    if res: 
      #print(res[0])
      kvp2[res[0]] = kvp[key].copy()
    else:
      kvp2[new_key] = kvp[key].copy()
  return kvp2

key_val = read_GI("inquirerbasic_py.csv")

```

```python
# for each player get tokens from sentence
def get_player_tokens(filename, eth):
  tok_list = []
  count = 0
  with open(filename) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    hiv4 = ps.HIV4()
    next(csv_reader)
    for row in csv_reader:
      if row[3] == eth:  
        count +=1   
        text = row[2]
        tokens = hiv4.tokenize(text)
        for tok in tokens:
          tok_list.append(tok)
  return tok_list

white_tokens = get_player_tokens('dataset_2021_weka.csv', 'W')
non_white_tokens = get_player_tokens('dataset_2021_weka.csv', 'NW')
```

```python
# match tokens to keys and save values (i.e the sentiment)
import operator

def get_GI_senti(tokens, gi_dict):
  counts = dict()
  w_count = 0
  for tok in tokens:
    for words in gi_dict:
      if tok == words:
        w_count += 1
        #print(f"token: {tok}, sentiment: {gi_dict[tok]}.")
        for senti in gi_dict[tok]:
          if senti in counts:
            counts[senti] += 1
          else:
            counts[senti] = 1
  #counts = dict(sorted(counts.items(),  key=operator.itemgetter(0),reverse=True))
  return counts, w_count

white_senti, w_word_c = get_GI_senti(white_tokens, key_val)
non_white_senti, nw_word_c = get_GI_senti(non_white_tokens, key_val)
```

```python
# get list and most popular sentiments for white and non-white
print(dict(sorted(white_senti.items(),  key=operator.itemgetter(1),reverse=True)))
sum(white_senti.values())
```

```python
def get_percentage(diction, total):
  x = {}
  #print(total)
  for k, v in diction.items():
    pct = v / total
    pct = pct * 100
    #print(k, pct)
    x[k] = round(pct, 2)
  return x

white_percent = get_percentage(white_senti, w_word_c)
non_white_percent = get_percentage(non_white_senti, nw_word_c)
```

```python
print(dict(sorted(white_percent.items(),  key=operator.itemgetter(1),reverse=True)))
```

```python
def create_csv(d1,d2):
  comb = [d1,d2]
  d = {}
  diff = []
  for k in white_percent.keys():
    d[k] = tuple(d[k] for d in comb)
  for k in d:
    v1 = d[k][0]
    v2 = d[k][1]
    top = v1-v2
    bottom = (v1+v2)/2
    rel_diff = (top / bottom) * 100
    diff.append(round(rel_diff, 2))
  kv_list = [[key, *val] for key, val in d.items()]
  # print(kv_list)
  with open('Enquirer_comparison_v2.csv', 'w', encoding='UTF8') as f:
      writer = csv.writer(f)
      header = ["Sentiment", "% of words with sentiment in white reports", "% of words with sentiment in non-white reports", "relative % difference"]
      writer.writerow(header)
      for i in range(0, len(kv_list)):
        kv_list[i].append(diff[i])
      writer.writerows(kv_list)

create_csv(white_percent, non_white_percent)
```
