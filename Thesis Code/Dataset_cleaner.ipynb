```python
import csv
```

```python
def get_reports(file):
  reports = []
  with open(file) as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)
    for row in csvreader:
      new = row[2].strip()
      print(new)
      reports.append(new)
  return reports

player_reports = get_reports("dataset_2021.csv")
```

```python
# creating dataset where all sentences are on one line, removing "\n" characters
def strip_reports(reps):
  length = len(reps)
  rep_stripped = []
  for i in range(0, length):
    new = reps[i].replace("\n", "")
    #print(new)
    rep_stripped.append(new)
  return rep_stripped
    

player_rep_cleaned = strip_reports(player_reports)
```

```python
# creating dataset where all sentences are on one line, removing "\n" characters
def create_weka_friendly_csv(file, reports):
  names = []
  positions = []
  ethnicity = []
  header = ["ID", "Position", "Scouting Report", "Race"]
  with open(file) as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)
    for row in csvreader:
      names.append(row[0])
      positions.append(row[1])
      ethnicity.append(row[3])
  with open("dataset_2021_weka.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    for i in range(0, len(names)):
      data = [i+1, positions[i], reports[i], ethnicity[i]]
      writer.writerow(data)
  return names, positions, ethnicity

names,pos,eth = create_weka_friendly_csv("dataset_2021.csv", player_rep_cleaned)
```
